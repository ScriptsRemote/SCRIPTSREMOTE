```python
import leafmap
import rasterio
import fiona
import os
```

```python
image = '/content/drive/MyDrive/MBA/img_rgb_pratica.tif'
src = rasterio.open(image)

m = leafmap.Map()
m.add_raster(src,indexes= [4, 3, 2], vmin= 0.01, vmax= 0.1, layer_name="RGB")
m
```

```python
mask = mask = m.user_roi
mask
```

```python
output = "clip.tif"
```

```python
try:
    leafmap.clip_image(image, mask, output)
except Exception as e:
    print(e)
```

```python
try:
    m.add_raster(output, indexes= [4, 3, 2], vmin= 0.01, vmax= 0.2, layer_name="RGB Clip")
except Exception as e:
    print(e)
```
