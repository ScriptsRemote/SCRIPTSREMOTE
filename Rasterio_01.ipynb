# **RASTERIO**

Neste exemplo, estamos usando a função show() da biblioteca rasterio.plot para realizar composições de bandas do GeoTIFF. Esta função usa uma projeção cartográfica do arquivo GeoTIFF para plotar o raster em suas coordenadas geográficas originais.

```python
##Instalação das bibliotecas
!pip install rasterio
```

```python
##importando as bibliotecas
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
```

Você pode personalizar o código para plotar outras bandas ou ajustar os parâmetros do gráfico, como a escala de cores ou a orientação. Verifique a documentação da biblioteca rasterio para obter mais informações sobre como personalizar o código para suas necessidades específicas.

```python
# abrir o arquivo GeoTIFF
with rasterio.open('/content/drive/MyDrive/TF/sentinel2_POA.tif') as src:   #caminho/do/arquivo.tif
    # ler as bandas rgb
    # normalizar as bandas entre 0 e 1
    src_rgb = src.read([4,3,2])/ 10000.0
    src_rnb = src.read([4,8,2])/ 10000.0
    src_ngb = src.read([8,3,2])/ 10000.0

    # plotar a composição RGB
    fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))
    # O valor 2 é um fator de escala que é multiplicado a cada pixel
    # da imagem para ajustar a intensidade da cor.
    # A escala de valores em um GeoTIFF geralmente é muito
    # maior do que a escala de cores usada para exibir a imagem na tela.
    ##Eixo 0
    show(src_rgb*2.5, ax=axs[0], transform=src.transform, title ='Image-bands 4,3,2')
    ##Eixo 1
    show(src_rnb*2.5, ax=axs[1],transform=src.transform, title = 'Image-bands 4,8,2')
    ##Eixo 2
    show(src_ngb*2.5, ax=axs[2],transform=src.transform, title = 'Image-bands 8,3,2')
    #Visualizar
    plt.show()
```
