<font color='orange'><b><font size='5'>Como baixar essa imagem para o meu PC?</font></b></font>

* Para realizar o download vamos usar a seguinte função

* Definir uma região de interesse (neste exemplo vamos começar com uma geometria)

* Clique aqui e acesse o Code Editor para Definir nossa geometria -> [Code Editor](https://code.earthengine.google.com/)

* Na barra de ferramentas à esquerda da tela, selecione a opção "Geometry" para criar uma nova geometria.

```python
##Região de interesse
roi = ee.Geometry.Polygon(
        [[[-42.91436270308333, -11.734190125622279],
          [-42.91436270308333, -14.409332274113611],
          [-39.98101309370833, -14.409332274113611],
          [-39.98101309370833, -11.734190125622279]]])

##Função para download
task = ee.batch.Export.image.toDrive(image=srtm,  # um objeto ee.Image.
                                     region=roi,  # um objeto ee.Geometry.
                                     description='Srtm_bahia',
                                     folder='MINICURSO_GEE',
                                     fileNamePrefix='Srtm_bahia',
                                     scale=30,
                                     maxPixels=1e13,
                                     crs='EPSG:4674')
##Inicie o Download
task.start()
```

```python
##Verifique o status
task.status()
```

<font color='orange'><b><font size='5'>Vamos extrair estatísticas da nossa região?</font></b></font>

* Iremos usar redutores do GEE
* Iremos usar a biblioteca pandas -> [Pandas](https://pandas.pydata.org/docs/getting_started/index.html)
* Iremos usar a biblioteca matplotlib -> [Matplotib](https://matplotlib.org/stable/index.html)

```python
##Primeiro precisamos criar nossos redutores
##Os redutores reduzem a média, mediana, mínimo e por ai vai
redutores = ee.Reducer.mean().combine(**{
    'reducer2': ee.Reducer.min(), 
    'sharedInputs': True}).combine(**{
        'reducer2': ee.Reducer.max(),
        'sharedInputs': True})

##Use a combinação para extrair os valore
estatistica_zonais = srtm.reduceRegion(**{
  'reducer': redutores,
  'geometry': roi,
  'scale':30,
  'bestEffort': True,
  'maxPixels':1e13,
  'tileScale':4
});

print(estatistica_zonais.getInfo());
```

```python
##importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##nossos dados estão como um dicionário
stats_list = ee.List(estatistica_zonais).getInfo()
##Extrair os valores dos items
data = list(stats_list.items())
##Para isso iremos cria um dataframe 
df = pd.DataFrame(data, columns=["elevação", "valor"])
df
```

```python
##Criando nosso gráfico 
# Extrair dados do dataframe
values = df["valor"]
labels = df["elevação"]

# Criar figura com tamanho 8 x 6 polegadas
plt.figure(figsize=(8, 6))

# Criar gráfico de colunas com os dados
plt.bar(labels, values, color="blue")

# Adicionar rótulos e título ao gráfico
plt.xlabel("Elevação")
plt.ylabel("Valor")
plt.legend(["Elevação (m)"])
plt.title("Gráfico de Colunas")

##Salvar gráfico
plt.savefig('Gráfico_barra.png')

# Mostrar o gráfico na tela
plt.show()
```

```python
##Importar biblioteca 
import rasterio as rio 
from rasterio.plot import show
```

```python
##Abrir nossa imagem 

##Precisaremos conectar nosso drive
filename = "/content/drive/MyDrive/MINICURSO_GEE/Srtm_bahia.tif"
raster = rio.open(filename)
```

```python
##Verificando os metadados
raster.meta
```

```python
##Obter os dados do raster e as informações de sua extensão:
data = raster.read(1)
extent = raster.bounds
```

```python
# Plotar o gráfico com a escala "terrain"
fig, ax = plt.subplots(1,1, figsize=(14,5))

# Plotar o raster com o rasterio.show() e integração com o plt
img = plt.imshow(data, cmap="terrain")  # Criar imagem vazia para a barra de legenda
plt.colorbar(img, label="Elevação (m)")  # Usar a imagem vazia como argumento
##Por fim o raster 
show(raster, ax=ax, transform=raster.transform, cmap="terrain")


# Personalizar rótulos e título do gráfico
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Mapa de Elevação SRTM")

##Inserindo grids
ax.grid(color='0.95', linestyle='dashed', linewidth=0.5, alpha=0.5)

##Plotar box de informações
ax.text(0.02,0.05,'CGS LAT/LONG SIRGAS 2000''\n Fonte:GEE \n Autor: Christhian Cunha',
         transform=ax.transAxes, fontsize= 8, bbox = dict(facecolor = 'white'))

##Vamos adicionar o norte
x, y, arrow_length = 0.95, 0.99, 0.1
##Usando uma anotação de x e y
ax.annotate('N', xy=(x, y), xytext=(x, y-arrow_length),
            arrowprops=dict(facecolor='black', width=5, headwidth=15),
            ha='center', va='center', fontsize=20,
            xycoords=ax.transAxes)

##Salvando a figura como jpeg
plt.savefig('Mapa de Elevação.jpg',dpi=300)

# Mostrar o gráfico na tela
plt.show()
```
